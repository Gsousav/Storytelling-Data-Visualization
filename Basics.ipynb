## Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
drop_index = euro_to_dollar[euro_to_dollar['US_dollar'] == '-'].index
euro_to_dollar.drop(drop_index, inplace=True)
euro_to_dollar.loc[:, 'US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

## Rolling Mean

Now I'll generate a line plot to visualize the evolution of the euro-dollar exchange rate
I'll use the rolling mean because it will show a smoother, easier to interpret, graph. Also, the data-ink ratio will be lower

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

### Coming up with an Idea

## Comparison: Exchange Rate of the US Dollar to the Euro

We will compare the exchange rate of the US dollar to the euro during the following significant economic crises in Europe over the last 20 years:

1. Global Financial Crisis (2007-2008)
3. Eurozone Crisis (2011-2013)
5. COVID-19 Pandemic (2020-2021)

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline  
```

```python
euro_to_dollar['year'] = euro_to_dollar['Time'].dt.year
euro_to_dollar['year'].value_counts().sort_index()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(5, 8))
axes = [ax1, ax2, ax3]

for ax in axes:
    ax.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'], color='#98bf64', alpha=0.35)
    ax.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
    ax.tick_params(top=0, left=0) 
    ax.grid(visible=False)

    
    for location in ['left', 'right', 'top', 'bottom']:
        ax.spines[location].set_visible(False)
        
global_financial_crisis = euro_to_dollar[(euro_to_dollar['year'] == 2007) | (euro_to_dollar['year'] == 2008)]
ax1.plot(global_financial_crisis['Time'], global_financial_crisis['rolling_mean'], linewidth=2.5, color='#607b3d')
gfc_time = global_financial_crisis['Time'].iloc[-1]  
ax1.text(gfc_time, ax1.get_ylim()[1], 'Global Crisis', fontsize=10, verticalalignment='top')

eurozone_crisis = euro_to_dollar[(euro_to_dollar['year'] == 2011) | (euro_to_dollar['year'] == 2012) | (euro_to_dollar['year'] == 2013)]
ax2.plot(eurozone_crisis['Time'], eurozone_crisis['rolling_mean'], linewidth=2.5, color='#607b3d')
ax2.set_xticklabels([])
ezc_time = eurozone_crisis['Time'].iloc[-1]
ax2.text(ezc_time, ax2.get_ylim()[1], 'Eurozone Crisis', fontsize=10, verticalalignment='top')

covid = euro_to_dollar[(euro_to_dollar['year'] == 2020) | (euro_to_dollar['year'] == 2021)]
ax3.plot(covid['Time'], covid['rolling_mean'], linewidth=2.5, color='#607b3d')
ax3.set_xticklabels([])
covid_time = covid['Time'].iloc[-1]
ax3.text(covid_time, ax3.get_ylim()[1], 'COVID-19', fontsize=10, verticalalignment='bottom')

fig.suptitle('During Three Outlined Crises: Euro to Dollar Exchange Rate Averaged 1.20', fontsize=14, fontweight='bold')
fig.text(0.5, 0.94, 'Charting Fluctuations: Global Financial Crisis (2007-2009), Eurozone Crisis (2011-2013), and COVID-19 Pandemic (2020-2021)', fontsize=10, ha='center')

plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```
